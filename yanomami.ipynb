```python
import geopandas
```

```python
import matplotlib.pyplot as plt
```

```python
mapa_br = geopandas.read_file("BR_UF_2021.shp")
```

```python
mapa_br.plot(color='#edf0f2')
plt.show()
```

```python
mapa_br.crs
```

```python
homologada = geopandas.read_file("tis_poligonais_h.shp")
homologada = homologada.to_crs(epsg=4674)
```

```python
homologada.crs
```

```python
homologada.plot()
plt.show()
```

```python
nao_homologada = geopandas.read_file("tis_poligonais_nh.shp")
nao_homologada = nao_homologada.to_crs(epsg=4674)
```

```python
nao_homologada.plot()
plt.show()
```

```python
mineracao = geopandas.read_file("BRASIL.shp")
```

```python
mineracao.crs
```

```python
mineracao = mineracao.to_crs(epsg=4674)
mineracao['ANO'].max()
```

```python
mineracao.plot()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30, 30))
mapa_br.plot(ax=ax, color='#edf0f2')
mineracao.plot(ax=ax, color='#52a5f2')
ax.set_axis_off()
plt.savefig('Mineracao_Br.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30, 30))
mapa_br.plot(ax=ax, color='#edf0f2')
homologada.plot(ax=ax, color='#03315c')
ax.set_axis_off()
plt.savefig('Homologacao_Br.png')
plt.show()
```

```python
mapa_rr = mapa_br[(mapa_br['NM_UF']=='Roraima')]
mapa_rr
```

```python
yanomami = homologada[homologada['etnia_nome']=='Yanommi']
```

```python
mineracao_rr = mineracao[(mineracao['UF']=='RR')]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
mapa_rr.plot(ax=ax, color='#edf0f2')
yanomami.plot(ax=ax, color='#03315c')
mineracao_rr.plot(ax=ax, color='#52a5f2', alpha=0.8)
ax.set_axis_off()
plt.savefig('Mineracao_Yanomami.png')
plt.show()
```
