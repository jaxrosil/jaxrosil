```python
import pandas as pd


df = pd.read_csv("arquivo.csv")

df_ibirapuera = df.query("Identificacao=='IBIRAPUERA'")

df_ibirapuera = df.groupby("Ano").sum()
df_ibirapuera.reset_index(inplace=True)
df_ibirapuera = df_ibirapuera[["Ano", "Num pessoas empregadas", "Orcamento total da Obra R$"]]

df_ibirapuera_qtde = df.groupby("Ano").count()
df_ibirapuera_qtde.reset_index(inplace=True)
df_ibirapuera_qtde = df_ibirapuera_qtde[["Ano", "Identificacao"]]
df_ibirapuera_qtde.rename(columns={'Identificacao': 'Num locacoes'},inplace=True)

df_soma = df.groupby("Ano").sum()
df_soma.reset_index(inplace=True)
df_soma = df_soma[["Ano", "Num pessoas empregadas", "Orcamento total da Obra R$"]]

df_quantidade = df.groupby(["Ano"]).count()
df_quantidade.reset_index(inplace=True)
df_quantidade = df_quantidade[["Ano", "Identificacao"]]
df_quantidade.rename(columns={'Identificacao': 'Num locacoes'},inplace=True)

from dash import Dash, html, dcc, Input, Output
import plotly.express as px
from jupyter_dash import JupyterDash

#import pandas as pd

#app = Dash(__name__)
app = JupyterDash(__name__)

#importante para o deploy
#server = app.server

#df = pd.read_excel("Vendas.xlsx")

# criando o gráfico
fig = px.line(df_soma, x="Ano", y="Orcamento total da Obra R$")
opcoes = ["Orcamento total da Obra R$ geral", "Num pessoas empregadas geral", "Num locacoes geral","Orcamento total da Obra R$ Ibirapuera", "Num pessoas empregadas Ibirapuera", "Num locacoes Ibirapuera"]


app.layout = html.Div(children=[
    html.H1(children='Situação das locações SP Film Commission antes das concessões privadas'),
    html.H2(children='Gráficos com parâmeros orçamento, núm de empregados e núm de locações de 2016-2021'),

    dcc.Dropdown(opcoes, value="Orcamento total da Obra R$ geral", id='parametros'),

    dcc.Graph(
        id='grafico',
        figure=fig
    )
])

@app.callback(
    Output('grafico', 'figure'),
    Input('parametros', 'value')
)
def update_output(value):
    if value == "Orcamento total da Obra R$ geral":
      fig = px.line(df_soma, x="Ano", y="Orcamento total da Obra R$")
    elif value == "Num pessoas empregadas geral":
      fig = px.line(df_soma, x="Ano", y="Num pessoas empregadas")
    elif value == "Num locacoes geral":
      fig = px.line(df_quantidade, x="Ano", y="Num locacoes")
    elif value == "Orcamento total da Obra R$ Ibirapuera":
      fig = px.line(df_ibirapuera, x="Ano", y="Orcamento total da Obra R$")
    elif value == "Num pessoas empregadas Ibirapuera":
      fig = px.line(df_ibirapuera, x="Ano", y="Num pessoas empregadas")
    elif value == "Num locacoes Ibirapuera":
      fig = px.line(df_ibirapuera_qtde, x="Ano", y="Num locacoes")

    return fig
```

```python
app.run_server(mode="inline")
```
